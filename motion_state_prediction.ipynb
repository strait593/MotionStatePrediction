#### Determine the time domain features for each motion state

```python
#extract features from df
#normalize data
axes = ['accelerometer_X', 'accelerometer_Y','accelerometer_Z']
df[axes] = (df[axes] - df[axes].mean() / df[axes].std())

print(df[axes])
```

```python
#windowing parameters
windows_size = 200
step_size = 100
features = []

#extract features
for i in range(0, len(df) - windows_size, step_size):
    window = df.iloc[i: i + windows_size]

    activity = window['activity'].mode()
    win_features = {'activity': activity}

    for axis in axes:
        win_features[f'{axis}_mean'] = window[axis].mean()
        win_features[f'{axis}_std'] = window[axis].std()
        win_features[f'{axis}_range'] = window[axis].max() - window[axis].min()
        win_features[f'{axis}_rms'] = np.sqrt(np.mean(window[axis]**2))
    
    win_features['SMA'] = (window[axes].abs().sum(axis=1).mean())

    features.append(win_features)
    
#create a separate dataframe for features
features_df = pd.DataFrame(features)
features_df
```
